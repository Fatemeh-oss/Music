```python
# Comparing top 20 artists hotness
df_name = df.groupby(["artist.name"], axis=0).mean()
df_name_sort = df_name.sort_values(by=["artist.hotttnesss"], ascending = False, axis=0).head(20)
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
df_name_sort["artist.hotttnesss"].plot(kind="barh", figsize=(10,6), color=colors_list).set(xlabel="Artist Hotness", ylabel = "Artist Name", title="Top 20 Artists Hotness")
```

```python
# Comparing top 20 songs hotness
df_song= df.groupby(["artist.name"], axis=0).mean()
df_song_sort = df_song.sort_values(by=["song.hotttnesss"], axis=0, ascending=False).head(30)
df_song_sort["song.hotttnesss"].plot(kind="barh", figsize=(15,10)).set(xlabel="Song Hotness", ylabel="Artist Name", title="Top 20 Songs Hotness")
plt.show()
```

```python
#Comparing song duration and loudness based on artists terms
df_sort = df.sort_values(["song.duration","song.loudness"], axis=0, ascending=False).tail(20)
x=df_sort["song.duration"]
y=df_sort["artist.terms"]
z=df_sort["song.loudness"]
fig = plt.figure(figsize=(14,6))
ax =fig.add_axes([0,0,1,1])
colors_list1 = ['gold']
colors_list2=['lightskyblue']
ax.scatter(y,x, color=colors_list1)
ax.scatter(y,z, color=colors_list2)
ax.legend(labels = ("song.duration","song.loudness"), loc="upper right")
ax.set_xlabel("Artists Terms")
ax.set_ylabel("Number")
ax.set_title("Comparing Song Duration and Loudness")
plt.show()
```

```python
#Comparing Artist & Song Hotness of Top10 Artists
df_grupby = df.groupby(["artist.name"], axis=0).mean()
df_top10 = df_grupby.sort_values(by="artist.hotttnesss", ascending = False, axis=0).head(10)
df_top10 = df_top10[["artist.hotttnesss","song.hotttnesss"]].head(10)
df_top10
```

```python
#Making Song Duration To 3 Categories for PiePlot
df['binned']=pd.cut(x=df['song.duration'], bins=[1.044440,176.038730,223.059140,276.375060,22050.000000])
```

```python
df_duration = df.groupby(["binned"], axis=0).sum()
colors_list = ['gold', 'lightcoral', 'lightskyblue', 'pink']
df_duration["song.bars_start"].plot(kind="pie", figsize = (10,6), autopct="%0.0f%%",labels=None, colors=colors_list, radius =1).set(title = "Comparing Percentage of Song Bars Start based on Song Duration Categories")
plt.legend(labels = df_duration.index, loc="upper right")
```
